```python
import pandas as pd
import requests
import os
```

```python
import requests
from bs4 import BeautifulSoup

# Send a GET request to the website
url = "https://lenr-canr.org/DetailOnly.htm"
response = requests.get(url)

# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the <a> tags that contain the paper titles and download links
paper_links = soup.find_all('a', href=True, text=True)

# Create a list to store the download links
download_links = []

# Loop through each paper link and extract the download link
for link in paper_links:
    # Get the download link
    download_link = link['href']

    # Add the download link to the list
    download_links.append(download_link)

# Print the list of download links
print(download_links)

```

```python
# Create a pandas DataFrame from the list of download links
df = pd.DataFrame({'Download Links': download_links})

# Write the DataFrame to an Excel file
df.to_excel('download_links.xlsx', index=False)
```

```python
# Find all the parent elements of the links that contain the serial numbers and paper titles
paper_parents = soup.find_all('p', {'class': 'papertitle'})

# Create a list to store the paper titles and download links
papers = []

# Loop through each parent element and extract the serial number, paper title, and download link
for parent in paper_parents:
    # Get the serial number
    serial_number = parent.find('a').text.strip()

    # Get the paper title
    paper_title = parent.contents[-1].strip()

    # Get the download link
    download_link = parent.find('a')['href']

    # Add the paper details to the list
    papers.append({'Serial Number': serial_number, 'Paper Title': paper_title, 'Download Link': download_link})

# Create a pandas DataFrame from the list of paper details
df = pd.DataFrame(papers)
print(df)
# Write the DataFrame to an Excel file
#df.to_excel('papers.xlsx', index=False)
```

```python
# Load the Excel file using pandas
df = pd.read_excel('download_links.xlsx')

# Set the directory to save the downloaded files
download_dir = 'data'
os.makedirs(download_dir, exist_ok=True)

# Loop through each download link and download the file
for index, row in df.iterrows():
    download_link = row['Download Links']
    filename = download_link.split('/')[-1]
    file_path = os.path.join(download_dir, filename)

    # Check if the file has already been downloaded
    if os.path.exists(file_path):
        print(f"{filename} already exists, skipping...")
        continue

    # Download the file using requests
    print(f"Downloading {filename}...")
    try:
        response = requests.get(download_link, stream=True)
        response.raise_for_status()
        with open(file_path, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8192):
                f.write(chunk)
    except requests.exceptions.RequestException as e:
        print(f"Error downloading {filename}: {str(e)}")
```
